## Importing data from a csv file and creating a horizontal bar graph  as well as showing some data cleaning skills<br>

### From the stack overflow survey, this data consist of developer known programming languages sorted by responder ID  

```python
langs = pd.read_csv('./Datasets/data.csv')
```

```python
langs.head(12)
```

### Before creating the graph, first we have to cleanv our data. To do this I first wanted to analyze the data, from the dataframe I called the series "LanguagesWorkedWith" and looked at the first five entries. After doing this i had a better idea of how the data was created and changed the formatting of the series but turning each row iteration into a list. 

```python
langs['LanguagesWorkedWith'].head()
```

```python
langs['LanguagesWorkedWith'] = langs['LanguagesWorkedWith'].str.split(';')
```

```python
langs['LanguagesWorkedWith'].head()
```

### Now that I have a list of all the known languages, i imported from the collections package counter. Now that I have the counter method, which returns a list of tuples containing the language name and their associated counter.  Finding the exact number of people who know each individual language is easier to determine. 

```python
language_counter = Counter()
for i in langs['LanguagesWorkedWith']:
    language_counter.update(i)

most_common = language_counter.most_common(15)
print(most_common)
```

```python
most_common[0][0]
```

### Finishing the data cleaning process by creating two indiviual list called, "languages" and "Popularity". The purpose of this is to more easily set my axis in the bar graph

```python
languages = []
popularity = []

for i in most_common:
    languages.append(i[0])
    popularity.append(i[1])
```

### Creating the Horizontal bar chart 

```python
# adjusting figure setting 
plt.figure(figsize = (5,4),dpi = 300)

# Descending list sort 
languages.reverse()
popularity.reverse()

# Creating bar chart 
plt.barh(languages,popularity)

# Basic graph essentials 
plt.title('Top 15 popular languages, According to 2019 Stack Overflow survay')
plt.xlabel('Number of developer ')
plt.xticks(rotation = 90)

# Testing purposes to see available styles which can be used 
# print(plt.style.available)

#Applying desired style 
plt.style.use('seaborn-notebook')

# saving the figure
plt.savefig('Popular_languages_HorizontalBarChart.png',bbox_inches='tight',dpi=100)

# showing graph
plt.show()
```
