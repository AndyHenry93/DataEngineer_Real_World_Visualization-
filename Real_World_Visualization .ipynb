# Real World Data Analyst Visualizations

The purpose of this project is to take real world datasets and apply python tools like numpy, pandas, and matplotlibs to create visaul graphs to help the understanding of the data.<br>
In this project I used two different datasets which include gas prices in USD from multiple countries in different regions. The other is a list of soccer players in the game FIFA. 

## Importing all the necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line Graph

```python
gas = pd.read_csv('./matplotlib_tutorial-master/gas_prices.csv')

gas
```

```python
# graph sizing 
fig = plt.figure(figsize =(10,5), dpi = 300)

# tesing with three gas prices
# plt.plot(gas['Year'],gas['USA'], label = 'USA', marker = 'o')
# plt.plot(gas['Year'],gas['Canada'], label = 'Canada', marker = 'o')
# plt.plot(gas['Year'],gas['South Korea'], label = 'South Korea', marker = 'o')

# Separate countries by Region
oceania =  ['Australia',]
north_america =  ['Canada','USA']
eu =  ['France','Germany','Italy','UK']
asia = ['Japan','South Korea','']
central_america = ['Mexico']

# plotting all countries
for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'],gas[country], label = country, marker = 'o')
        
# # plotting countries specified by region
# for country in gas:
#     if country in central_america:
#         plt.plot(gas['Year'],gas[country], label = country, marker = 'o')
        
# Creating Title
plt.title('Country gas prices in (USD), from 1990-2008',fontdict = {'fontsize':20})

# Creating axis labels and legend 
plt.xlabel('Years', fontdict = {'fontsize':13})
plt.ylabel('USD', fontdict = {'fontsize':13})
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)

# correcting x-axis ticks
x_axis = (gas['Year'][::3])
plt.xticks(x_axis)

# saving the figure
plt.savefig('Country_gas_prices.png',bbox_inches='tight',dpi=100)

# Showing Graph
plt.show()
```

```python
fifa = pd.read_csv('../matplotlib_tutorial-master/fifa_data.csv')
fifa.head()
```

## Histograms

```python
# X-axis parameters 
bins = [40,50,60,70,80,90,100]

# creating the histogram 
plt.hist(fifa.Overall,bins, color = '#38B2E1', edgecolor ='black' )
plt.xticks(bins)

# Labeling the graph
plt.title('FIFA 2018 player ranking, based on skill level')
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
# saving the figure
plt.savefig('FIFA_ranking.png',bbox_inches='tight',dpi=100)
plt.show()
```

## Pie Chart

```python
fifa.head()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

print(left)
print(right)
```

```python
plt.pie([left,right], labels = ['Left foot','Right foot'], colors = ['#DC3150', '#1FD2AC'],shadow = True, autopct ='%.2f %%')
plt.title('Foot preference of FIFA players')
# saving the figure
plt.savefig('FIFA_dominateFoot.png',bbox_inches='tight',dpi=100)
plt.show()
```
